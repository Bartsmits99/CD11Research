```python
import pickle

import numpy as np
import pandas as pd
import random

import plotly.graph_objects as go

from flightObj import Flight;

```

```python
with open('data/[31].pkl', 'rb') as f:
    se_31 = pickle.load(f)
with open('data/[31_T_A320].pkl', 'rb') as f:
    se_31_T_A320 = pickle.load(f)
with open('data/[31_O].pkl', 'rb') as f:
    se_31_O = pickle.load(f)
with open('data/[31_I].pkl', 'rb') as f:
    se_31_I = pickle.load(f)
with open('data/[31_E].pkl', 'rb') as f:
    se_31_E = pickle.load(f)
with open('data/[31_M].pkl', 'rb') as f:
    se_31_M = pickle.load(f)
with open('data/[31_L].pkl', 'rb') as f:
    se_31_L = pickle.load(f)
```

```python
slices = {'A320':[se_31_T_A320, '[31T_A320]'],
          'All':[se_31,'[31]'],
          'Outbound':[se_31_O, '[31_O]'],
          'Inbound':[se_31_I, '[31_I]'],
         }
```

```python
cs = 'Inbound'  #choose your slice here :)

flights = slices[cs][0]
```

```python
sfig = go.Figure()

random.seed(314159)
sflight = random.sample(list(flights),1)[0]

ds = sflight.ds
dn = sflight.dn
do = sflight.do

sfig.add_trace(go.Scatter3d(x=ds['X'], y=ds['Y'], z=ds['MODE_C'], text=sflight.call, name='TRUE           ' + str(len(ds))))
sfig.add_trace(go.Scatter3d(x=dn['X'], y=dn['Y'], z=dn['MODE_C'], text=sflight.call, name='PCHIP          ' + str(len(dn))))
sfig.add_trace(go.Scatter3d(x=do['X'], y=do['Y'], z=do['MODE_C'], text=sflight.call, name='PCHIP_RDP \t' + str(len(do))))

sfig.update_traces(mode='lines+markers', marker_line_width=0, marker_size=1.5, line_width=1.5)

sfig.update_layout(title=sflight.call, autosize=True,
        scene=dict(aspectratio = dict( x=5, y=5, z=1), aspectmode = 'manual'))

sfig.show()

print(sflight.call)
sfig.write_html(('figs/' + sflight.call + '_vis2_single.html'))
```

```python
fig = go.Figure()

for flight in flights:
    do = flight.do
    
    do["FLIGHT_ID"] = do["FLIGHT_ID"].astype(str)
    do["TPLUS"] = do["TPLUS"].astype(str)
    
    if flight.ftype == 'OUTBOUND':
        txt = 'T+ ' + do['TPLUS'] + '<br>' +'Outbound at ' + flight.aptime + ' for ' + flight.dest + '<br>' +'Type: ' + flight.actype
    elif flight.ftype == 'INBOUND':
        txt = 'T+ ' + do['TPLUS'] + '<br>' + 'Inbound at ' + flight.aptime + ' from ' + flight.adep + '<br>' +'Type:  ' + flight.actype
    else:
        txt = flight.ftype
        
    trace = go.Scatter3d(x=do['X'], y=do['Y'], z=do['MODE_C'],
                         name=flight.call,
                         text=txt
                         
                         )
    
    fig.add_trace(trace)
    
fig.update_traces(mode='lines+markers', marker_line_width=0, marker_size=1.5, line_width=1.5)

fig.update_layout(
    title='Slice: ' + slices[cs][1],
    
    autosize=True,
 
   
    scene=dict(
        aspectratio = dict( x=5, y=5, z=1),
        aspectmode = 'manual'
    )    
)

fig.show()

fig.write_html('figs/' + slices[cs][1] + '_vis2_multi.html')

```
